```
%matplotlib inline
# Important for using matplotlib w/ Ipython Notebook
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import graphviz as gv

import random
import numpy as np

import sys
import os
```

```
def GraphIt(V, E, w={}, D=True, C={}):
    'Plots a graph'
    G = gv.Digraph(filename='graph', format='png') if D else gv.Graph(filename='graph', format='png')
        
    for v in V:
        G.node(str(v))
    for u in E:
        for v in E.get(u, []):
            if not D and u>v: continue
            if (u,v) in C or ((v,u) in C and not D):
                G.edge(str(u),str(v), label = str(w.get((u,v), '')), color='blue')
            else:
                G.edge(str(u),str(v), label = str(w.get((u,v), '')))
    G.render()
    img=mpimg.imread('graph.png')

    fig, ax = plt.subplots(frameon=False)
    fig.set_size_inches(7,7)
    imgplot = plt.imshow(img)

    ax.axis('off')
    plt.box(on=None) 
    plt.show()  
    print ("Vertex: Edge List")
    for u in V:
        print (u, ": ", E.get(u, []))
    
def PrintTreeWalk(G, x):
  if x!=None:
    G.node(str(x.key))
    if x.left != None:
      G.node(str(x.left.key))
      G.edge(str(x.key),str(x.left.key), label = '')
      PrintTreeWalk(G, x.left)
    if x.right != None:
      G.node(str(x.right.key))
      G.edge(str(x.key),str(x.right.key), label = '')
      PrintTreeWalk(G, x.right)
    
def PrintTree(T):
    'Plots a tree'
    G = gv.Graph(filename='graph', format='png')
    
    PrintTreeWalk(G, T.root)
    
    G.render()
    img=mpimg.imread('graph.png')

    fig, ax = plt.subplots(frameon=False)
    fig.set_size_inches(7,7)
    imgplot = plt.imshow(img)

    ax.axis('off')
    plt.box(on=None) 
    plt.show()  
    print ("Vertex: Edge List")
    for u in V:
        print (u, ": ", E.get(u, []))
```

```
def GraphFromEdges(N, edgeList):
    V = list(range(1,N+1))
    E = {}
    for a,b in edgeList:
        e = E.get(a,[])
        e.append(b)
        e.sort()
        E[a] = e
    return V, E
  
graphEdges = [(8,7), (1, 10), (2, 1),
              (3, 9), (4, 1), (5, 3),
              (5, 6), (5, 9), (5, 10),
              (6, 2), (6, 3), (7, 1),
              (7, 2), (7, 9), (7, 10),
              (9, 1), (9, 4), (9, 10),
              (8,2)
             ]
V, E = GraphFromEdges(10, graphEdges)
print (V, E)
GraphIt(V, E)
```

```
class Tree:
  pass

class Node:
  def __init__(self, key):
    self.p=None;
    self.key = key
    self.left = None
    self.right = None
    
def TreeInsert(T, z):
  y=None
  x=T.root
  while (x != None):
    y = x
    if z.key < x.key:
      x = x.left
    else:
      x = x.right
  z.p = y
  if y==None:
    T.root = z
  elif z.key < y.key:
    y.left = z
  else:
    y.right = z
    
def InorderTreeWalk(x):
  if x!=None:
    InorderTreeWalk(x.left)
    print (x.key, end=" " )
    InorderTreeWalk(x.right)
    
```
